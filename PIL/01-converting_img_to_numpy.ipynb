```python
from PIL import Image
img = Image.open('./img/ojisan.png')
print(type(img))
img.show()
```

```python
import numpy as np
img_np = np.asarray(img)
print(type(img_np))

# showing numpy array as img

back_to_img = Image.fromarray(img_np)
back_to_img.show()

back_to_img = Image.fromarray(img_np, 'RGB')
back_to_img.show()
```
