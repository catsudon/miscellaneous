```python
from PIL import Image, ImageOps
import numpy as np

bg = Image.open('./img/city.jpg')
img = Image.open('./img/ojisan.png')
w,h = bg.size
print(bg.size)
```

```python
ojisan = ImageOps.contain(img, bg.size)
ojisan.show()
print(img.size , "converted to ", ojisan.size)
```

```python
data = np.zeros((h, w, 4), dtype=np.uint8)
data[0:h , 0:w] = [0, 0, 0, 0]
final = Image.fromarray(data)
final.paste(ojisan, ((w-ojisan.size[0])//2,0), ojisan)
final.show()
final.save('./img/resized_ojisan.png')
```
