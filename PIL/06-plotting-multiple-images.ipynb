```python
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

ojisan = Image.open('./img/ojisan.png')
ojisan = np.array(ojisan)

plt.subplot(121)
plt.imshow(ojisan)
plt.subplot(122)
plt.imshow(ojisan,cmap="pink") # can't use cmap for RBG val anyway
plt.axis("on")
```

```python
from skimage.feature import canny
from skimage.color import rgb2gray, rgba2rgb

ojisan = rgba2rgb(ojisan)
ojisan = rgb2gray(ojisan)
fig,axes = plt.subplots(nrows=6,ncols=8,sharex=True,sharey=True,figsize=(28,28))
ax = axes.ravel()
for i in range(48):
    ax[i].imshow(canny(ojisan,sigma=i/10.0),cmap="gray")
    ax[i].set_title("canny sigma = " + str(i/10.0))
    ax[i].axis("off")
```
