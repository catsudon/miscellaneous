# For pasting png on top of png

#### (images need to have the same size)

```python
bg = Image.open('./img/city.png')
ojisan = Image.open('./img/resized_ojisan.png')
w,h = bg.size
```

```python
blended = Image.blend(bg.convert('L'), ojisan.convert('L'), alpha=0.4)
blended.show()
```

```python
final = Image.new("RGBA", ojisan.size)
final = Image.alpha_composite(final, bg)
final = Image.alpha_composite(final, ojisan)
final.show()
```
