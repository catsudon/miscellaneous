```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np

bg = Image.open('../img/city.jpg')
ojisan = Image.open('../img/ojisan.png')
w,h = bg.size
```

```python
data = np.zeros((h, w, 4), dtype=np.uint8)
data[0:h , 0:w] = [0, 0, 0, 0]
trans = Image.fromarray(data)
trans.paste(ojisan, ((w-ojisan.size[0])//2,150), ojisan)
trans.show()
```

```python
blended = Image.blend(bg.convert('L'), trans.convert('L'), alpha=0.65)
blended.show()
```

```python
font_type = ImageFont.truetype('./SOV_SongKhrueang.ttf' , 70)

I1 = ImageDraw.Draw(blended)
I1.text((130, 36), "", fill=150, font=font_type)
blended.show()
```

```python
blended.save('output.jpg')
```
